```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"]=(12,6)
df = pd.read_csv("COVID-19.csv.zip")
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
plt.figure(figsize=(14,6))

plt.plot(df['ObservationDate'], df['Confirmed'])

plt.title("Daily Confirmed COVID-19 Cases")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")

plt.xticks(rotation=60)

plt.show()
```

```python
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])

plt.figure(figsize=(14,6))

plt.plot(df['ObservationDate'], df['Deaths'], color='red')

plt.title("COVID-19 Death Trend")
plt.xlabel("Date")
plt.ylabel("Deaths")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(14,6))

plt.plot(df['ObservationDate'], df['Recovered'], color='green')

plt.title("COVID-19 Recovery Trend")
plt.xlabel("Date")
plt.ylabel("Recovered")

plt.xticks(rotation=45)

plt.show()
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(15,7))

plt.plot(df['ObservationDate'], df['Confirmed'], label='Confirmed')

plt.plot(df['ObservationDate'], df['Deaths'], label='Deaths')

plt.plot(df['ObservationDate'], df['Recovered'], label='Recovered')

plt.legend()

plt.title("COVID-19 Trends")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="Blues")

plt.title("Correlation Heatmap")

plt.show()
```
